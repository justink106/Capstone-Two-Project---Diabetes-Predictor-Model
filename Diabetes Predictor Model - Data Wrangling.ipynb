```python
# Import pandas, numpy, seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Loading the Dataset

```python
diabetes_df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
```

```python
# Now checking to see if there are any missing values in this dataset by column

missing = pd.concat([diabetes_df.isnull().sum(), 100 * diabetes_df.isnull().mean()], axis = 1)
missing.columns = ['count', '%']
missing.sort_values(by = 'count', ascending = False)
```

The above shows me that I am working with a dataset with no missing values.

```python
# Looking at the individual count for people who have diabetes, pre-diabetes, or no diabetes

diabetes_count = diabetes_df['Diabetes_012'].value_counts().reset_index()
print(diabetes_count)
```

```python
diabetes_df
```

Most individuals in this dataset do not have Diabetes

```python
# No need to have the data type for this dataset to be floats as no decimals are needed
# Changing the data type from float to integer

diabetes_df = diabetes_df.astype(int)
```

```python
# Checking to make sure the data type is integer now

print(diabetes_df.dtypes)
```

```python
# Sorting the dataframe using the 'Diabetes_012' column
diabetes_df.index.name = 'Index'
diabetes_df = diabetes_df.sort_values(["Diabetes_012", "Index"])
```
